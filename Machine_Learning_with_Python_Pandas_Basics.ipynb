## Machine Learning with Python
Working with music.csv and automating predictions
@4:33:00 in the Youtube https://youtu.be/_uQrJ0TkZlc

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

music_data = pd.read_csv('music.csv')
music_data
```

### Input Set & Output Set: divide into sets

We have Age & Gender as input (which we cannot change or manupulate, received exclusively). And we have Genre, which can differ from person to person, regardless of wheather they are male/female Or different age groups.

Output dataset is the one that we will ask our programme to predict.

```python
X = music_data.drop(columns=['genre']) # Creating Input DataSet
y = music_data['genre'] # Output DataSet, which will be used for predictions

model = DecisionTreeClassifier() # Initializing a Model
model.fit(X, y) # Offering input/output parameters
predictions = model.predict([ [21, 1], [22, 0] ]) # Asking to predict some missing elements (21 years old Male & 22 years old Female)
predictions # Model predicted HipHop & Dance
```

### Calculating the Accuracy

We need to divide the dataset into two part; one for Training and one for Testing. Using 2 samples as done above is not enough. Generally use 70-80% data for Training and 20-30% testing.

```python
X = music_data.drop(columns=['genre']) 
y = music_data['genre']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)


model = DecisionTreeClassifier() 
model.fit(X_train, y_train) 
predictions = model.predict(X_test)

score = accuracy_score(y_test, predictions)
score # shows accuracy 1.0 for 100%, and then changes to .75/.50/.25 if we keep running
```

### Persist and Load Models

Once creating and training a model, we may want to save it in a file. Next time then we can simply load the model and ask for predictions.

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
import joblib

music_data = pd.read_csv('music.csv')
X = music_data.drop(columns=['genre']) 
y = music_data['genre']

model = DecisionTreeClassifier() 
model.fit(X, y) 

joblib.dump(model, 'music-recommender.joblib') # Saved joblib file in the directory
```

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
import joblib


model = joblib.load('music-recommender.joblib') # Now loading the model, and predicting the below data
predictions = model.predict([ [21, 1] ])
predictions
```

### Visualizing a Decision Tree

This will create a .dot file in the directory. If we open VSCode and install extension Graphviz (dot) by Stephanvs, and drag/drop the .dot file inside VSCode. On the top right corner we will be 3 dots, from there we can preview our decision tree (A decision tree made by our model)

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree

music_data = pd.read_csv('music.csv')
X = music_data.drop(columns=['genre']) 
y = music_data['genre']

model = DecisionTreeClassifier() 
model.fit(X, y)

tree.export_graphviz(model, out_file='music-recommender.dot',
                     feature_names=['age', 'gender'],
                     class_names=sorted(y.unique()),
                     label='all',
                     rounded=True,
                     filled=True)


```
